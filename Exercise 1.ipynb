```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# пользовательская переменная для хранения размера шрифта
fsize=10
# настройка типа шрифта на рисунке с помощью изменения
# записей в словаре rcParams из модуля pyplot
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'
# настройка размера шрифта в различных частях рисунка
# в заголовке:
plt.rcParams['axes.titlesize'] = 12
# в подписях осей:
plt.rcParams['axes.labelsize'] = 10
# в подписях меток на осях:
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
# в легенде рисунка:
plt.rcParams['legend.fontsize'] = fsize
# массив координат – 50 точек в диапазоне [0, 10]
x = np.linspace(0.0, 2*math.pi, 50)
# создаём окно рисунка.
# Для дальнейшей работы рисунок ассоциируется с переменной
# fig
fig=plt.figure()
# добавляем панель (оси координат) с именем ax в окно fig.
# в дальнейшем настройка осей производится через обращение к переменным ax1, ax2, ax3, ax4.
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3= fig.add_subplot(2,2,3)
ax4= fig.add_subplot(2,2,4)
ax1.set_title('(а)')
ax2.set_title('(б)')
ax3.set_title('(в)')
ax4.set_title('(г)')
# график синуса: сплошная (-) чёрная линия.
ax1.plot(x, np.sin(x), 'k-')
# график косинуса: пунктирной (:) черной линией
ax2.plot(x, np.cos(x), 'k:')
# график косинуса в квадрате: штриховой (--) черной линией
ax3.plot(x, (np.cos(x))**2.0, 'k--')
# график функции f(x)=x^0.15: сплошной (-) серой линией 
ax4.plot(x, (x)**0.15, '-', color = 'grey')
# диапазон отображаемых значений по оси х
ax1.set_xlim(0, 2*math.pi)
ax2.set_xlim(0, 2*math.pi)
ax3.set_xlim(0, 2*math.pi)
ax4.set_xlim(0, 2*math.pi)
# диапазон отображаемых значений по оси y
ax1.set_ylim(-1, 1)
ax2.set_ylim(-1, 1)
ax3.set_ylim(0, 1)
ax4.set_ylim(0, 2)
# подпись по оси x
ax1.set_xlabel(r'$x$')
ax2.set_xlabel(r'$x$')
ax3.set_xlabel(r'$x$')
ax4.set_xlabel(r'$x$')
# подпись по оси y
ax1.set_ylabel(r'$f(x)$')
ax2.set_ylabel(r'$f(x)$')
ax3.set_ylabel(r'$f(x)$')
ax4.set_ylabel(r'$f(x)$')
# сохраняем в файл с именем fig2 типа PNG с разрешением
# 300 точек на дюйм
# (dpi – dots per inch), с альбомной ориентацией
fig.savefig("fig2.png", orientation='landscape', dpi=300)
```
